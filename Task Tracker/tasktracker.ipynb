```python
import json
import tkinter as tk

tasks = []
checkbox_vars = []
states = []

def añadir_tarea(task):
    if task == "":
        return
    else:
        tasks.append(task)
        task_entry.delete(0, tk.END)
        var = tk.BooleanVar()  # Create a new BooleanVar for each checkbox
        checkbox = tk.Checkbutton(window, text=task, variable=var)
        checkbox.pack(anchor='w', padx=10)
        checkbox_vars.append(var)  # Store the BooleanVar in the list
        return

def load():
    with open("tasks.json", "r", encoding="UTF-8") as f, with open("done_tasks.txt", "a", encoding="UTF-8") as fp:
        if f.read() == "":
            return
        else:
            f.seek(0)
            dic = json.load(f)
            eliminar = []
            for key in dic:
                if dic[key]:
                    fp.write(" - "+str(key).capitalize()+"\n")
                    eliminar.append(key)
            for x in eliminar:
                dic.pop(x)
            for key in dic:
                tasks.append(key)
                var = tk.BooleanVar()
                var.set(dic[key])
                checkbox = tk.Checkbutton(window, text=key, variable=var)
                checkbox.pack(anchor='w', padx=10)
                checkbox_vars.append(var)
                states.append(dic[key])
    return

window = tk.Tk()
window.geometry("400x400")
window.grid()
window.title("Task Manager")

title = tk.Label(window, text="Task Manager", font=("Arial", 20))
title.pack(anchor='w', padx=10)

task_entry = tk.Entry(window)
task_entry.pack(anchor='w', padx=10, pady=10)
submit_button = tk.Button(window, text="Add", font=("Helvetica", 10), command=lambda: añadir_tarea(task_entry.get()))
submit_button.place(x=150, y=47)

load()

def print_checkbox_states():
    global states
    states = []
    for i, var in enumerate(checkbox_vars):
        state = True if var.get() else False
        states.append(state)
    print(states)
# Button to check the states of checkboxes
check_button = tk.Button(window, text="Save", command=print_checkbox_states)
check_button.place(x=10, y=350)

window.mainloop()

dic = {}
print(states)
f = open("tasks.json", "w", encoding="UTF-8")
for x in range(len(tasks)):
    dic[tasks[x]] = states[x]
print(dic)
json.dump(dic, f)
dic.clear()
f.close()
```
