```python
import json
import tkinter as tk

tasks = []
checkbox_vars = []

def a単adir_tarea(task):
    if not task:
        return
    tasks.append(task)
    task_entry.delete(0, tk.END)
    var = tk.BooleanVar()  # Create a BooleanVar for each checkbox
    checkbox = tk.Checkbutton(window, text=task, variable=var)
    checkbox.pack(anchor='w', padx=10)
    checkbox_vars.append(var)  # Store the BooleanVar in the list

def load():
    try:
        with open("tasks.json", "r", encoding="UTF-8") as f, open("done_tasks.txt", "a", encoding="UTF-8") as fp:
            if not f.read().strip():
                return
            f.seek(0)
            task_data = json.load(f)
            tasks_to_remove = []

            for task, done in task_data.items():
                if done:
                    fp.write(f" - {task.capitalize()}\n")
                    tasks_to_remove.append(task)

            # Remove completed tasks from the data
            for task in tasks_to_remove:
                del task_data[task]

            # Add remaining tasks to the UI
            for task, done in task_data.items():
                tasks.append(task)
                var = tk.BooleanVar(value=done)
                checkbox = tk.Checkbutton(window, text=task, variable=var)
                checkbox.pack(anchor='w', padx=10)
                checkbox_vars.append(var)
    except FileNotFoundError:
        pass  # If the file doesn't exist, do nothing

def save_tasks():
    task_states = {tasks[i]: var.get() for i, var in enumerate(checkbox_vars)}
    with open("tasks.json", "w", encoding="UTF-8") as f:
        json.dump(task_states, f)
    print("Tasks saved:", task_states)

# Create main window
window = tk.Tk()
window.geometry("400x400")
window.title("Task Manager")

# Title label
title = tk.Label(window, text="Task Manager", font=("Arial", 20))
title.pack(anchor='w', padx=10)

# Task entry and add button
task_entry = tk.Entry(window)
task_entry.pack(anchor='w', padx=10, pady=10)

submit_button = tk.Button(window, text="Add", font=("Helvetica", 10), command=lambda: a単adir_tarea(task_entry.get()))
submit_button.place(x=150, y=47)

# Load existing tasks
load()

# Save button
save_button = tk.Button(window, text="Save", command=save_tasks)
save_button.place(x=10, y=350)

# Run the application
window.mainloop()
```
